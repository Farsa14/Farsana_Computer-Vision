# Data Augumentation

    In computer vision, we deal with different problems that are mainly related to images. 
    Image classification is one of the tasks in computer vision where machines are now capable of doing classification more accurately than humans
    But the main drawback of image classification "large volumes of data". 

## Major Problem while training Deep learning

    Overfitting 
    underfitting  

## Regularization Technique

    Reduce the overfitting in the model
    
    If we get good results in training data and poor results in unseen data (test data, validation data) then it is framed
    as an overfitting problem
    

## Are you lacking with data??

     Solved by data augmentation ==> we create new data based on modifications of our existing data

## Pre-Processing or Post-Processing

    Data augmentation is a preprocessing technique because we only work on the data to train our model. In this technique, we generate new instances of images by 
    
    Horizontal flip
    Vertical flip
    Rotation
    Zoom in
    Zoom out
    Cropping
    Color variations
    Shear
    
So, whenever the training lacks the image dataset, by using augmentation we can create thousands of images to train the model 

##  Method : ImageDataGenerator()

    It accepts a batch of images which is used in training
    Takes the batch and applies series of random transformation to each image
    Replaces the original image with randomly transformed batch
    Trains the Neural Neteork on radomly transformed batch data and not on original data
    

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator,img_to_array, load_img
import matplotlib.pyplot as plt
```

```python
pred_gen = ImageDataGenerator(
    rotation_range = 80,
    width_shift_range = 0.4,
    height_shift_range = 0.4,
    shear_range = 0.4,
    zoom_range = 0.4,
    horizontal_flip = True,
    fill_mode = 'nearest')
```

```python
pred_img = load_img(r'C:\Users\DELL\Pictures\predator.jpg')

plt.imshow(pred_img)
plt.show()
```

```python
pred_img_array = img_to_array(pred_img)
```

```python
pred_img_array = pred_img_array.reshape((1,) + pred_img_array.shape)
```

```python
i = 0
for batch in pred_gen.flow(pred_img_array, save_to_dir= 'Predator', save_prefix='predator', save_format='jpg'):
    i += 1
    if i >= 14:
        break
```
