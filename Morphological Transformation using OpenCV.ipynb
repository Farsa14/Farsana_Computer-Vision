# Morphological Transformation

Morphological transformation is basically some simple operations performed on a image. It need 2 inputs, one is original image and second one is kernel

There are main two operations in Morphological Transformation:

1.  Erosion 
2.  dilation 


```python
#To install opencv
#pip install opencv-python
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread(r"C:\Users\DELL\Pictures\npl.jpg")
plt.imshow(img)
plt.show()
```

```python
img.shape 
```

```python
kernel = np.ones((3,3))
kernel
```

### Erosion

    we are just omitting the boundaries of the front image, we are thinning the object
    loss of pixels
    Note: Hppens on the edges not from middle

```python
erosion = cv2.erode(img,kernel,iterations=3)
plt.imshow(erosion)
plt.show()
```

### Dilation
    It’s actually the reverse process of Erosion, we are just going to thick the boundaries of an image. 
    Gain of pixels

```python
dilation = cv2.dilate(img,kernel)
plt.imshow(dilation)
plt.show()
```


Whats the ulitimate goal of these opearions???
removing noise from images


```python
img3 = cv2.imread(r"C:\Users\DELL\Pictures\npl2.jpg")
plt.imshow(img3)
plt.show()
```

```python
#Here we have many straches, how can we tune here 
kernel = np.ones((5,5))
noise_rem = cv2.erode(img3,kernel,iterations=1)
plt.imshow(noise_rem)
plt.title("Image after removing noise")
plt.show()
```

```python
#mix erosion and dialtion based on position
kernel1 = np.ones((7,7))
img4 = img3.copy()
img4[:,:200] =cv2.dilate(cv2.erode(img4[:,:200],kernel1),kernel1)
plt.imshow(img4)
plt.show()
```

### Opening
It is actually an operation that consist of an erosion followed by a dialation

```python
opening = cv2.morphologyEx(img,cv2.MORPH_OPEN,kernel)
plt.imshow(opening)
plt.show()
```

### Closing
 opposite to Opening, means dilation followed by erosion 

```python
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
plt.imshow(closing)
plt.show()
```

### Gradient
Difference between the dialtion and erosion of the image

```python
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)
plt.imshow(gradient)
plt.show()                           
```

### Tophat
show the difference between the input image and the opening image 

```python
tophat = cv2.morphologyEx(img, cv2.MORPH_TOPHAT, kernel)
plt.imshow(tophat)
plt.show() 
```

### Blackhat 
show the difference between the input image and the closing image 

```python
blackhat = cv2.morphologyEx(img, cv2.MORPH_BLACKHAT, kernel)
plt.imshow(blackhat)
plt.show() 
```
